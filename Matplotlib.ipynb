# SubPlot



```python
x = data['Length1']
y1= data['Weight']
y2 = data["Height"]
y3 = data["Width"]
y4= data["Length2"]
```

```python
#Graph 1
plt.subplot(2,3,1)  # 2= No.of rows, 3 = No of columns , 1 = Figure no
plt.plot(x,y1)
plt.title("Graph 1")
plt.xlabel("Length")
plt.ylabel("Weight")

# Graph 2
plt.subplot(2,3,2)
plt.plot(x,y2)
plt.title("Graph 2")
plt.xlabel("Length")
plt.ylabel("Height")

# Graph 3
plt.subplot(2,3,3)
plt.plot(x,y3)
plt.title("Graph 3")
plt.xlabel("Length")
plt.ylabel("Width")

# Graph 4
plt.subplot(2,3,4)
plt.plot(x,y4)
plt.title("Graph 4", loc = 'left')
plt.xlabel("Length")
plt.ylabel("Length")

plt.suptitle("Fish_Dataset") # suptitle = supertitle
plt.show
```

# Cmap

```python
# Graph1
plt.subplot(1,2,1)
plt.scatter(x,y4)


#Graph2
plt.subplot(1,2,2)
plt.scatter(x,y1, color= 'red',cmap = 'viridis')

plt.show()
```

```python
colors= np.random.randint(100, size=(159))
size = np.random.randint(1000,size=(159))
```

```python
plt.subplot(1,2,1)
plt.scatter(x, y4,c=colors,cmap='viridis',s =size,alpha= 0.5)


plt.subplot(1,2,2)
plt.scatter(x,y1,c=colors,cmap='viridis',s=size, alpha = 0.5)
plt.colorbar()
plt.show()
```
